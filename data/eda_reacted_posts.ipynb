```python
import matplotlib.pyplot as plt
from tqdm import tqdm
import sweetviz as sv
import pandas as pd
import numpy as np
import json
import os
```

```python
with open('raw/reacted_posts.json') as json_file:
    posts = pd.DataFrame(json.load(json_file))
```

```python
print(posts['ml_language'].isnull().values.any())
print(posts['post_id'].isnull().values.any())
print(posts['ml_interests'].isnull().values.any())
```

# Inspecting the keys included under 'content' attribute

```python
for index in range(len(posts)):
    print(posts.loc[index,'content'].keys())
```

# Splitting key-value pairs in 'content' into distinct attributes

```python
for index in tqdm(range(len(posts))):
    content = posts.loc[index, 'content']
    
    if 'description' in content.keys():
        posts.loc[index, 'content_description'] = content['description']
    
    if 'caption' in content.keys():
        posts.loc[index, 'content_caption'] = content['caption']
        
    if 'title' in content.keys():
        posts.loc[index, 'content_title'] = content['title']
```

```python
posts = posts.fillna("")
posts
```
