```python
from matplotlib_venn import venn3, venn3_circles
import matplotlib.pyplot as plt
import sweetviz as sv
import pandas as pd
import numpy as np
import json
import os
```

```python
# Opening JSON file
with open('raw/user_info.json') as json_file:
    data = json.load(json_file)
```

# Record Count

```python
user_profile_df = pd.DataFrame(data)
user_profile_df
```

# Unique Users Count

```python
print(user_profile_df['user_id'].isnull().values.any())
len(user_profile_df['user_id'].unique())
```

# Missing Count in Languages

```python
print(user_profile_df['language'].isnull().values.any())
print(user_profile_df['language'].isnull().sum())
```

# Language Values Distribution



```python
user_profile_expanded = user_profile_df.explode('language')
report = sv.analyze(user_profile_expanded)
report.show_html()
```

# INSPECTING THE TOP 3 MOST FREQUENT LANGUAGES

# Filtering out users that view "hi" content

```python
hi_users = user_profile_expanded[user_profile_expanded['language']=="hi"]['user_id'].tolist()
hi_users
```

# Language Distribution among "hi" users



```python
hi_users_df = user_profile_df[user_profile_df['user_id'].isin(hi_users)].reset_index(drop=True)
hi_users_expanded = hi_users_df.explode('language')
report = sv.analyze(hi_users_expanded)
report.show_html()
```

# Filtering out users that view "mr" content

```python
mr_users = user_profile_expanded[user_profile_expanded['language']=="mr"]['user_id'].tolist()
mr_users
```

# Language Distribution among "mr" users



```python
mr_users_df = user_profile_df[user_profile_df['user_id'].isin(mr_users)].reset_index(drop=True)
mr_users_expanded = mr_users_df.explode('language')
report = sv.analyze(mr_users_expanded)
report.show_html()
```

# Filtering out users that view "en" content

```python
en_users = user_profile_expanded[user_profile_expanded['language']=="en"]['user_id'].tolist()
en_users
```

# Language Distribution among "en" users



```python
en_users_df = user_profile_df[user_profile_df['user_id'].isin(en_users)].reset_index(drop=True)
en_users_expanded = en_users_df.explode('language')
report = sv.analyze(en_users_expanded)
report.show_html()
```

# Obtaining users from the rest of the languages

```python
te_users = user_profile_expanded[user_profile_expanded['language']=="te"]['user_id'].tolist()
ta_users = user_profile_expanded[user_profile_expanded['language']=="ta"]['user_id'].tolist()
bn_users = user_profile_expanded[user_profile_expanded['language']=="bn"]['user_id'].tolist()
gu_users = user_profile_expanded[user_profile_expanded['language']=="gu"]['user_id'].tolist()
kn_users = user_profile_expanded[user_profile_expanded['language']=="kn"]['user_id'].tolist()
ml_users = user_profile_expanded[user_profile_expanded['language']=="ml"]['user_id'].tolist()
rest_users = user_profile_expanded[user_profile_expanded['language']==""]['user_id'].tolist()
```

```python
def view_overlap(d1,d2,d3,l1,l2,l3):
    venn3(
         [set(d1), set(d2), set(d3)],
         set_colors=('#3E64AF', '#3EAF5D', '#D74E3B'), 
         set_labels = (l1, l2, l3),
         alpha=0.75
)
    venn3_circles([set(d1), set(d2), set(d3)], lw=0.7)
    plt.show()
```

```python
view_overlap(d1=hi_users, d2=mr_users, d3=en_users, l1="hi", l2="mr", l3="en")
```

```python
index_labels = {
    0:"hi",
    1:"mr",
    2:"en",
    3:"te",
    4:"ta",
    5:"bn",
    6:"gu",
    7:"kn",
    8:"ml",
    9:"<empty>"
}

all_users = [set(hi_users), set(mr_users),
             set(en_users), set(te_users),
             set(ta_users), set(bn_users),
             set(gu_users), set(kn_users),
             set(ml_users), set(rest_users)]
             
for index in range(len(all_users)):
    print(f"Size of {index_labels[index]} = {len(all_users[index])}")
    for inner_index in range(index+1, len(all_users)):
        print(f"---Size of {index_labels[inner_index]} = {len(all_users[inner_index])}")
        print(f"-----Intersection Size between {index_labels[index]} and {index_labels[inner_index]} = {len(all_users[index].intersection(all_users[inner_index]))}")
```
